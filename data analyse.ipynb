```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

pd.options.display.max_columns=100

#2.Récupération du dataset avec Pandas
titanic = pd.read_csv("https://bit.ly/2C2Tup2", sep=';')
head_data = titanic.head()
tail_data = titanic.tail()
info_data = titanic.info()
describe_data = titanic.describe()
(head_data, head_data, tail_data, info_data)

#3.Types de données
titanic["sex"] = titanic["sex"].astype("category")
titanic["pclass"] = titanic["pclass"].astype("category")
titanic["survived"] = titanic["survived"].astype("category")
titanic["embarked"] = titanic["embarked"].astype("category")
sex_counts = titanic["sex"].value_counts()
pclass_counts = titanic["pclass"].value_counts()
survived_counts = titanic["survived"].value_counts()
embarked_counts = titanic["embarked"].value_counts()
titanic["age"] = titanic["age"].str.replace(',', '.').astype(float)
(sex_counts, pclass_counts, survived_counts, embarked_counts)

#4.Données manquantes et hors-format
print(titanic.info())
titanic["age"] = titanic["age"].astype(str).str.replace(',', '.').fillna('28').astype('float32')
dexieme_info = titanic.info()
dexieme_describe = titanic.describe()
(dexieme_info, dexieme_describe)

#5.Supprimer les colonnes "inutiles"
titanic = titanic.drop(['body', 'ticket', 'fare', 'cabin', 'name', 'boat', 'home.dest'], axis = 1)
troisieme_info = titanic.info()
troisieme_info

#6.Graphique "camembert"
plt.figure(figsize = (4, 4))
plt.title = "Survived Distribution"
plt.pie(
   titanic["survived"].value_counts(),
   labels = ["survived", "not survived"],
   autopct = "%1.1f%%",
   startangle=90
)
plt.show()

plt.figure(figsize = (4, 4))
plt.title = "Pclass Distribution"
plt.pie(
   titanic["pclass"].value_counts(),
   labels = ["3 class", "2 class", "1 class"],
   autopct = "%1.1f%%",
   startangle = 90
)
plt.show()

#7.BoxPlot
#age boxplot
plt.figure(figsize = (4, 4))
plt.title = "Boxplot of Age"
titanic["age"].plot.box()
plt.show()
#pclass et age de boxplot
plt.figure(figsize = (4, 4))
plt.title = "Age and Class Distribution"
sb.boxplot(x = "pclass", y = "age", data = titanic)
plt.xlabel("Class")
plt.ylabel("Age")
plt.show()

#8.Histogrammes
plt.figure(figsize=(4, 4))
plt.hist(titanic["age"].dropna(), bins=8, edgecolor="pink")
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
